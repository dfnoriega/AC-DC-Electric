```python
#imports libraries that will be utilized
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame, Series
import requests
import scipy.stats as sts
import scipy as spy
import scipy.integrate as integrate
import scipy.optimize as opt
import random
import numpy.linalg as lin
import pickle as pkl
```

```python
#Loads Information provided and creates elements with additional basic information

#Loads profiles from file provided
#Values in [MW]
profiles=pd.read_csv('Profiles.csv',sep=",")
profiles.columns=["Year","Month","Day","Hour","Load","Wind","Solar","NetLoad"]

#Loads Generators Stack
#Values in [MW],[hr],[%]
GenStack=pd.read_csv('GenStack.csv',sep=",")
GenStack.columns=["Gen","FuelType","Pmin","PMax","minRuntime","minDowntime","HeatRate","ForcedOutageRate","RepairTime"]
pd.options.mode.chained_assignment = None
for gen in range (0,len(GenStack)):
    GenStack['PMax'][gen]=GenStack['PMax'][gen]*(1-(GenStack['ForcedOutageRate'][gen]))

#Loads Fuel Prices
#Values in [$/MMbtu]
FuelPrices=pd.read_csv('FuelPrices.csv',sep=",")
CoalP=np.float(FuelPrices['Coal'][0])
NatGP=np.float(FuelPrices['NaturalGas'][0])
OilP=np.float(FuelPrices['Oil'][0])

#Creates element for monthly averages of daily Solar Radiation
#Monthly averages Values in [kWh/m2/day]
#From https://pvwatts.nrel.gov/pvwatts.php
MonSolIrradiation=np.array([4.3,5.32,6.35,6.82,6.96,7.35,6.73,6.41,6.37,5.42,4.88,4.21])

#Creates element for monthly averages of Daylight and sunshine times
#Monthly averages hours per day [hr]
#From https://www.weather-us.com/en/colorado-usa/denver-climate#daylight_sunshine
DaylightHours=np.array([10,11,12,13,14,15,15,14,12,11,10,9])
SunshineHours=np.array([6.9,7.5,8.2,9.2,9.4,10.5,10.5,9.9,9.1,8,6.5,6.3])

#Assumed Efficiency for PV
ConvEff=0.15

periods=len(profiles['Load'])

#Creates elements that will be used to store variables for the optimization and results
status=pd.DataFrame(np.zeros((len(GenStack),4)))
status.columns=["Off","On","TimeOff","TimeOn"]
GenMix=np.zeros((len(GenStack),periods))
ShadowCost=np.zeros((1,periods))
Gaps=np.zeros((1,periods))
MaxGenVector=np.zeros((1,periods))
CostVector=np.zeros((1,periods))
ViolationsVector=np.zeros((1,periods))
```

```python
#Calculates area covered by the panels
#Peak Sun Conditions in [W/m2], Efficiency in [%], MW is the installed capacity desired
#Returns area in [m2]
def area(PeakSun,efficiency,MW):
    area = (MW*(10**6))/(PeakSun*efficiency)
    return area
```

```python
#Creates Solar Output Profile
#Takes average daily values and estimates output for each hour
#See slides for intuition on hourly values

HrSolarIrradiance=np.zeros((periods,1))
AreaInstalled=area(1000,ConvEff,1)
SolarOutput=np.zeros((periods,1))

#Solar Output in MW
for t in range (0,periods):
    month=profiles['Month'][t]-1
    DaylightTime=DaylightHours[month]
    DailyIrradiation=MonSolIrradiation[month]

    if profiles['Hour'][t] > (13.001-((DaylightTime)/2)):
        if profiles['Hour'][t] < (13.001+((DaylightTime)/2)):
            if SolarOutput[t,0]==0:
                SolarVector=np.zeros((1,DaylightTime))
                averageTime=(DaylightTime+1)/2
                TotalWeight=0
                for hr in range (0,DaylightTime):
                    Weight=1-(((averageTime-(hr+1))/DaylightTime)**2)
                    TotalWeight=TotalWeight+Weight
                    SolarVector[0,hr]=Weight
                for hr in range (0,DaylightTime):
                    Weight=SolarVector[0,hr]
                    HrSolarIrradiance[t+hr,0]=DailyIrradiation*Weight/TotalWeight
                    SolarOutput[t+hr,0]=(HrSolarIrradiance[t+hr,0]*AreaInstalled*ConvEff)/1000

SolarOutputProfile=pd.DataFrame(SolarOutput)
SolarOutputProfile.columns=['Unitary Solar Output (1MW Installed) [MW]']

#Exports Solar Output profile to an excel file
with open('SolarProfile1MW.xlsx', "wb+") as f:
    SolarOutputProfile.to_excel(f, index=False)
    f.close
```

```python
#Calculates the maximum generation capacity given generators that are running at a point in time
def MaxGeneration(status):
    Generation=0
    for gen in range (0,len(GenStack)):
        Generation=Generation+(GenStack['PMax'][gen]*status['On'][gen])
        
    return Generation
```

```python
#Function that updates cummulative values for the status of generators (Total hours On and Off)
def UpdateCummStatus(status0, status1):
    for gen in range (0, len(GenStack)):
        if status0['Off'][gen]==1:
            if status1['Off'][gen]==1:
                status1['TimeOff'][gen]=status0['TimeOff'][gen]+1
                status1['TimeOn'][gen]=0
            if status1['On'][gen]==1:
                status1['TimeOff'][gen]=0
                status1['TimeOn'][gen]=1
                status1['Off'][gen]=0
        if status0['On'][gen]==1:
            if status1['On'][gen]==1:
                status1['TimeOn'][gen]=status0['TimeOn'][gen]+1
                status1['TimeOff'][gen]=0
            if status1['Off'][gen]==1:
                status1['TimeOff'][gen]=1
                status1['TimeOn'][gen]=0
                status1['On'][gen]=0
    return status1
```

```python
#Function that updates current values for the status of generators
def UpdateStatus(status0,unitsOn):
    status1=status0
    for gen in range (0, len(GenStack)):
        if unitsOn[gen]==1:
            status1['On'][gen]=1
            status1['Off'][gen]=0
        if unitsOn[gen]==0:
            status1['Off'][gen]=1
            status1['On'][gen]=0
    return status1
```

```python
#Function that calculates the shadow cost of electricity given current generation mix
def marginalCost(status):
    marCost=0
    for gen in range (0,len(GenStack)):
        if status['On'][gen]==1:
            typ=GenStack['FuelType'][gen]
            if typ == "Coal":
                price=CoalP
            if typ == "Gas":
                price=NatGP
            if typ == "Oil":
                price=OilP
            
            marCostGen=GenStack['HeatRate'][gen]*price/1000
            
            if marCost<marCostGen:
                marCost=marCostGen
    
    return marCost
```

```python
#Function that resets values of status so that all generators are available
def resetStatus(status):
    status1=pd.DataFrame(np.zeros((len(GenStack),4)))
    status1.columns=["Off","On","TimeOff","TimeOn"]
    for gen in range (0,len(GenStack)):
        status1['Off'][gen]=1
        status1['TimeOff'][gen]=999
    return status1
```

```python
#Function that calculates cheapest generation mix to cover load
def GenProfile(status,Load,t):
    maxGen=MaxGeneration(status)
    unitsOn=np.array(status['On'])
    unitsOff=np.array(status['Off'])
    TimeOn=np.array(status['TimeOn'])
    TimeOff=np.array(status['TimeOff'])
    minDowntime=np.array(GenStack['minDowntime'])
    minRuntime=np.array(GenStack['minRuntime'])
    PMax=np.array(GenStack['PMax'])
    if Load>maxGen:
        generation=maxGen
        gen=0
        while Load>generation and gen<=13:
            if unitsOff[gen]==1:
                if TimeOff[gen]>=minDowntime[gen]:
                    unitsOn[gen]=1
                    generation=generation+PMax[gen]
            
            gen=gen+1
    
    if Load<maxGen:
        gen=len(GenStack)-1
        generation=maxGen
        while Load<generation and gen>=0:
            if unitsOn[gen]==1:
                if TimeOn[gen]>=minRuntime[gen]:
                    generation=generation-PMax[gen]
                    if Load<generation:
                        unitsOn[gen]=0
            gen=gen-1
    
    gap=0
    for gen in range (0,len(GenStack)-1):
        if unitsOn[gen+1]==1:
            if unitsOn[gen]==0:
                gap=gap+1

#The code below could be used for the full optimization, which includes look aheads.
#
#     sumviolations=0
#     v=t
#     while v<t+(np.max(minDowntime)*12) and v<periods:
#         sumviolations=sumviolations+ViolationsVector[0,v]
#         v=v+1
    
#     if sumviolations==0:
#         iteration=1
#         status1=UpdateCummStatus(status,UpdateStatus(status,unitsOn))
#         while gap>0 and iteration<4:
#             gen=len(GenStack)-1
#             while gen>=0:
#                 if unitsOn[gen]==1:
#                     if TimeOn[gen]>minRuntime[gen]:
#                         genit=0
#                         control=0
#                         rowGen=gen
#                         unitsOnAux=unitsOn
#                         unitsOnAux[gen]=0
#                         statusAux=UpdateCummStatus(status1,UpdateStatus(status1,unitsOnAux))
#                         while genit<rowGen and control==0:
#                             if unitsOnAux[genit]==0:
#                                 if TimeOff[genit]>minDowntime[genit]:
#                                     unitsOnAux[gen]=1
#                                     statusAux2=UpdateCummStatus(statusAux,UpdateStatus(status1,unitsOnAux))
#                                     maxGenAux=MaxGeneration(statusAux2)
#                                     if Load<maxGenAux:
#                                         unitsOn=unitsOnAux
#                                         gap=gap-1
#                                         control=1
#                             genit=genit+1

#                 gen=gen-1

#             iteration=iteration+1

    return unitsOn, gap
```

```python
#Function that selects generators for each period's demand and returns shadow cost of electricity
#Prints number of violations (i.e. periods when load > max generation capability)
def Optimization(Load):    
    statust=resetStatus(status)
    ViolationsVector=np.zeros((1,periods))
    for t in range (0,periods):
        unitsOn, gap = GenProfile(statust,Load[0,t],t)
        status1=UpdateCummStatus(statust,UpdateStatus(statust,unitsOn))
        marCost=marginalCost(status1)
        maxgen=MaxGeneration(status1)
        CostVector[0,t]=Load[0,t]*marCost
        MaxGenVector[0,t]=maxgen
        ShadowCost[0,t]=marCost
        GenMix[:,t]=unitsOn
        Gaps[0,t]=gap
        statust=status1

    for t in range (0,periods):
        if MaxGenVector[0,t]<Load[0,t]:
            ViolationsVector[0,t]=1

    print('Number of violations:',np.sum(ViolationsVector))
    AverageShadowCost=np.sum(CostVector)/np.sum(Load)
    
    return AverageShadowCost
```

```python
#Function that modifies net load according to Solar MWs installed
def NewNetLoad(SolarMW):
    Load0=profiles['NetLoad']
    Load1=np.zeros((1,periods))
    for t in range (0,periods):
        SolarProfile=SolarOutput[t,0]*SolarMW
        Load1[0,t]=Load0[t]-SolarProfile
        
    return Load1
```

```python
#Runs optimization for levels of solar penetration specified in the "Solar MW Vector"(line below)
SolarMWVector=[0,1.5,10,100,300,900,1500,3000]
AvShadowCostVector=np.zeros((1,len(SolarMWVector)))
TotalLoadVector=np.zeros((1,len(SolarMWVector)))

for penetration in range (0,len(SolarMWVector)):
    Load = NewNetLoad(SolarMWVector[penetration])
    AverageShadowCost=Optimization(Load)
    AvShadowCostVector[0,penetration]=AverageShadowCost
    TotalLoadVector[0,penetration]=np.sum(Load)
    
```

```python
#Shows results, plots normalized value of Solar MW
cases=len(SolarMWVector)
caseNormRes=np.zeros((1,cases))
for case in range (1,cases):
    normValue = (AvShadowCostVector[0,0]-AvShadowCostVector[0,case])/SolarMWVector[case]
    caseNormRes[0,case]=normValue

print('Levels of Penetration Considered:',SolarMWVector[:])
print('Average shadow cost of electricity for each case:',AvShadowCostVector)
print('Normalized Marginal Value of Solar PV:', caseNormRes[0,:])

plt.plot(SolarMWVector[:],caseNormRes[0,:])
```
